```python
import tellurium as te
r = te.loadSBMLModel('antagonist1.xml')
species = r.getFloatingSpeciesIds()
for s in species:
    r.resetAll()
    r.reset
    #r = te.loadSBMLModel('antagonist1.xml')
    sim = r.simulate(0,10,100, ['time',s])
    te.plotWithLegend(r,sim)
```
